```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from pycocotools.coco import COCO
from PIL import Image
import matplotlib.patches as patches
```

```python
ann_dir = "./annotations/instances_val2017.json"
coco = COCO(ann_dir)
```

```python
img_ids = coco.getImgIds(catIds=[2])
print(img_ids)
```

```python
for id in img_ids:
    ann_ids = coco.getAnnIds(imgIds=id, catIds=[2])
    print((len(ann_ids)))
```

```python
ann_ids = coco.getAnnIds(imgIds=343561, catIds=[2])
print((len(ann_ids)))
```

```python
anns = coco.loadAnns(ann_ids)
anns
```

```python
for ann in anns:
    print(ann['bbox'])
```

```python
img_id = 343561
img_path = "./images/000000343561.jpg"
image1 = Image.open(img_path)
image2 = Image.open(img_path)

fig, ax = plt.subplots()

for ann in anns:
    box = ann["bbox"]
    box_boundary = patches.Rectangle((box[0],box[1]),box[2],box[3], linewidth = 2, edgecolor = "blue", facecolor="none")
    ax.add_patch(box_boundary)

ax.imshow(image1)

```
